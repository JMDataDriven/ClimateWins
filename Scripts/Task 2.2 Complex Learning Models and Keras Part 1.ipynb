```python
# Create a dictionary to store the count of stations for each weather observation
station_counts = {}

for obs in weather_observations:
    # Select columns related to the current observation type
    columns = [col for col in df_filtered.columns if col.endswith(obs)]
    
    # Count the number of stations (i.e., the number of columns) for the current observation type
    station_counts[obs] = len(columns)

# Print the count of stations for each observation typed
print("Number of stations covered by each observation type:")
for obs, count in station_counts.items():
    print(f"{obs}: {count} stations")
```

```python
# Create a list of all unique station names in the dataset

stations_list = set([col.split('_')[0] for col in df_filtered_2.columns if '_' in col])
stations_list
```

```python
weather_types = ['cloud_cover', 'humidity', 'pressure']

missing_stations_by_weather_type = {}

for type in weather_types:
    # Select columns related to the current observation type
    columns = [col for col in df_filtered_2.columns if col.endswith(type)]
    
    # Extract station names by removing the observation type from the column names
    stations = set([col.replace(f'_{type}', '') for col in columns])
    
    # Identify stations that are in all_stations but missing from the current observation type
    missing_stations = stations_list - stations
    
    # Store the missing station names in the dictionary
    missing_stations_by_weather_type[type] = missing_stations

# Print the missing station names for each observation type
for type, missing_stations in missing_stations_by_weather_type.items():
    print(f"\nStations missing from {type}:")
    if missing_stations:
        for station in missing_stations:
            print(station)
    else:
        print("None")
```

```python
stations = {
0: 'BASEL',
1: 'BELGRADE',
2: 'BUDAPEST',
3: 'DEBILT',
4: 'DUSSELDORF',
5: 'HEATHROW',
6: 'KASSEL',
7: 'LJUBLJANA',
8: 'MAASTRICHT',
9: 'MADRID',
10: 'MUNCHENB',
11: 'OSLO',
12: 'SONNBLICK',
13: 'STOCKHOLM',
14: 'VALENTIA'

}
```

#### Attempt 1

```python
# Create a Keras layered model 
epochs = 8
batch_size = 16
n_hidden = 32

timesteps = len(X_train[0])
input_dim = len(X_train[0][0])
n_classes = len(y_train[0])

model = Sequential()
model.add(LSTM(n_hidden, input_shape=(timesteps, input_dim)))
model.add(Dropout(0.5))
model.add(Dense(n_classes, activation='softmax')) #Don't use relu here!
```

```python
model.compile(loss='categorical_crossentropy',
              optimizer='rmsprop',
              metrics=['accuracy'])
```

```python
model.fit(X_train,
          y_train,
          batch_size=batch_size,
          validation_data=(X_test, y_test),
          epochs=epochs)
```

```python
def confusion_matrix(Y_true, Y_pred):
    Y_true = pd.Series([stations[y] for y in np.argmax(Y_true, axis=-1)])
    Y_pred = pd.Series([stations[y] for y in np.argmax(Y_pred, axis=-1)])

    return pd.crosstab(Y_true, Y_pred, rownames=['True'], colnames=['predictions'])
```

```python
# Evaluate
print(confusion_matrix(y_test, model.predict(X_test)))
```

## CNN Model

```python
epochs = 10
batch_size = 512
n_hidden = 1024

timesteps = len(X_train[0])
input_dim = len(X_train[0][0])
n_classes = len(y_train[0])

model_5 = Sequential()
model_5.add(Conv1D(n_hidden, kernel_size=2, activation='relu', input_shape=(timesteps, input_dim)))
model_5.add(Dense(16, activation='relu'))
model_5.add(MaxPooling1D())
model_5.add(Flatten())
model_5.add(Dense(n_classes, activation='softmax')) #sigmoid
```

```python
model_5.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model_5.fit(X_train, y_train, batch_size=batch_size, epochs=epochs, verbose=2)
```

```python
# Evaluate
print(confusion_matrix(y_test, model_5.predict(X_test)))
```
